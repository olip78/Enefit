```python
def n_parameters(model):
    m = []
    for parameter in model.parameters():
        m.append(np.prod(list(parameter.shape)))
    print(f'# params = {sum(m)}')

    
def _plot(train_losses, val_losses, val_mae, val_mse, val_mae_real):
    fig, ax = plt.subplots(4, 1, figsize=(7, 13))

    ax[0].plot(range(len(train_losses)), train_losses)
    ax[0].plot(range(len(val_losses)), val_losses)
    ax[0].set_title('train / val loss')

    ax[1].plot(range(len(val_mse)), val_mse)
    ax[1].set_title('val MSE')
    
    ax[2].plot(range(len(val_mae)), val_mae)
    ax[2].set_title('val norn. MAE')
    
    ax[3].plot(range(len(val_mae_real)), val_mae_real)
    ax[3].set_title('val real MAE')


def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
hist_shifts=[2]

X, H, Y, _ = data_loader.generate(15, hist_shifts=hist_shifts)

# ----- model -------

encoder_embed_size = 64
decoder_embed_size = 64
encoder_output_length = len(hist_shifts)*24
encoder_output_size = [encoder_output_length, decoder_embed_size]
decoder_output_size = len(targets)*24

# encoder
input_shape = list(X.shape)

num_layers = 1
forward_expansion = 1
heads = 4
dropout_encoder = 0.3

encoder = Encoder(input_shape,
                  encoder_output_size,
                  encoder_embed_size,
                  num_layers, forward_expansion, heads, dropout_encoder, device).to(device)

# decoder
input_shape = list(H.shape)

num_layers = [1, 1]
heads = 1
dropout_decoder = 0.3
e_memory_len = encoder_output_length 

decoder = Decoder(input_shape,
                  decoder_embed_size,
                  e_memory_len,
                  decoder_output_size,
                  heads,
                  num_layers, forward_expansion, dropout_decoder, device).to(device)

model = Transformer(encoder, decoder, device)
n_parameters(model)

# ----- training -------

NUM_EPOCHS = 200
lr = 0.00005

mape = MeanAbsolutePercentageError().to(device)
mae = MeanAbsoluteError().to(device)
mse = MeanSquaredError().to(device)

loss_fn = torch.nn.MSELoss()
#loss_fn = torch.nn.L1Loss()

optimizer = torch.optim.Adam(model.parameters(), lr=lr)


train_losses = []

val_losses = []
val_mae = []
val_mse = []
val_mae_real = []

# training   
for j in range(NUM_EPOCHS):
    # train
    total_loss = 0

    t0 = 10
    t1 = 615
    train_steps = t1 - t0
    model.train()
    for b in data_blocks[t0: t1]:
        X, H, Y, units = data_loader.generate(b, hist_shifts=hist_shifts)
        optimizer.zero_grad()
        
        # model
        output = model(X.to(device), H.to(device))
        loss = loss_fn(output, Y.to(device))
        loss.backward()
        optimizer.step()
        total_loss += loss

    train_losses.append(total_loss.item() / train_steps)    
    
    # validation
    total_loss = 0
    total_mae = 0
    total_mse = 0
    total_mae_real = 0

    #val loop
    v0 = 615
    v1 = 635
    val_steps = v1 - v0
    model.eval()
    with torch.no_grad():
        for b in data_blocks[v0: v1]:
            X, H, Y, units = data_loader.generate(b, hist_shifts=hist_shifts)
            output = model(X.to(device), H.to(device))

            loss = loss_fn(output, Y.to(device))

            total_loss += loss
                
            total_mae += mae(output, Y.to(device))
            total_mse += mse(output, Y.to(device))
            total_mae_real += mae(dn.back_transform(output.detach().cpu().numpy(), units),
                                  dn.back_transform(Y.detach().cpu().numpy(), units))

        val_losses.append(total_loss.item() / val_steps)
        val_mae.append(total_mae.item() / val_steps)
        val_mse.append(total_mse.item() / val_steps)
        val_mae_real.append(total_mae_real.item() / val_steps)

    print(f'{j}: train loss: {train_losses[-1]:.5f}, val mse: {val_mse[-1]:.5f}, val mae: {val_mae[-1]:.5f}, real val mae: {val_mae_real[-1]:.2f}')

_plot(train_losses, val_losses, val_mae, val_mse, val_mae_real)
```
